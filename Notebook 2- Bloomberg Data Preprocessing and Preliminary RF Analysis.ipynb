```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

df['Dates'] = pd.to_datetime(df['Dates'])
df = df.sort_values('Dates')
```

```python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

train_start_date = pd.to_datetime('2015-01-01')
train_end_date = pd.to_datetime('2018-9-30')
validation_start_date = pd.to_datetime('2018-9-30')
validation_end_date = pd.to_datetime('2018-12-31')
test_start_date = pd.to_datetime('2019-1-1')
test_end_date = pd.to_datetime('2019-7-30')

df['target'] = df['3 Day Future Return'] > 0

train_data = df[(df['Dates'] >= train_start_date) & (df['Dates'] <= train_end_date)]
validation_data = df[(df['Dates'] >= validation_start_date) & (df['Dates'] <= validation_end_date)]
test_data = df[(df['Dates'] >= test_start_date) & (df['Dates'] <= test_end_date)]

features = ['Basic Earnings per Share', 'Last Price', 'Open Price', 'Volatility 10 Day', 
            'Volatility 30 Day', 'Volatility 90 Day', 'Volume', 'Price Earnings Ratio (P/E)',
            'Return on Common Equity', 'Total Debt to Total Equity']

best_accuracy = 0
best_params = {}

tscv = TimeSeriesSplit(n_splits=5)

n_estimators_options = [50, 100, 200]
max_depth_options = [3, 5, 7]
min_samples_split_options = [2, 5, 10]

for n_estimators in n_estimators_options:
    for max_depth in max_depth_options:
        for min_samples_split in min_samples_split_options:
            model = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth, min_samples_split=min_samples_split, random_state=42)
            temp_accuracy = []
            for train_index, val_index in tscv.split(train_data):
                X_train, X_val = train_data[features].iloc[train_index], train_data[features].iloc[val_index]
                y_train, y_val = train_data['target'].iloc[train_index], train_data['target'].iloc[val_index]

                model.fit(X_train, y_train)

                val_predictions = model.predict(X_val)
                val_accuracy = accuracy_score(y_val, val_predictions)
                temp_accuracy.append(val_accuracy)

            avg_accuracy = sum(temp_accuracy)/len(temp_accuracy)
            if avg_accuracy > best_accuracy:
                best_accuracy = avg_accuracy
                best_params = {'n_estimators': n_estimators, 'max_depth': max_depth, 'min_samples_split': min_samples_split}

best_model = RandomForestClassifier(n_estimators=best_params['n_estimators'], max_depth=best_params['max_depth'], min_samples_split=best_params['min_samples_split'], random_state=42)
best_model.fit(train_data[features], train_data['target'])

test_predictions = best_model.predict(test_data[features])
test_accuracy = accuracy_score(test_data['target'], test_predictions)

print(f'Best Parameters: {best_params}')
print(f'Test Accuracy: {test_accuracy}')
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

train_start_date = pd.to_datetime('2015-01-01')
train_end_date = pd.to_datetime('2018-12-31')
test_start_date = pd.to_datetime('2019-1-1')
test_end_date = pd.to_datetime('2019-7-30')
# Create target
df['target'] = df['3 Day Future Return'] > 0

# Split the data into train and test data
train_data = df[(df['Dates'] >= train_start_date) & (df['Dates'] <= train_end_date)]
test_data = df[(df['Dates'] >= test_start_date) & (df['Dates'] <= test_end_date)]

# Define the features used for the prediction
features = ['Basic Earnings per Share', 'Last Price', 'Open Price', 'Volatility 10 Day', 
            'Volatility 30 Day', 'Volatility 90 Day', 'Volume', 'Price Earnings Ratio (P/E)',
            'Return on Common Equity', 'Total Debt to Total Equity']

# Set up and train the model
model = RandomForestClassifier(n_estimators=200, max_depth=3, min_samples_split=2, random_state=42)
model.fit(train_data[features], train_data['target'])

# Make predictions on the test data
test_predictions = model.predict(test_data[features])

# Calculate the accuracy of the model on the test data
test_accuracy = accuracy_score(test_data['target'], test_predictions)

print(f'Test Accuracy: {test_accuracy}')
```

```python
from sklearn.metrics import accuracy_score
import numpy as np

# Assuming test_predictions is your array of predictions
test_data['predictions'] = test_predictions

# List to store results
results = []

# List of unique companies in the dataset
companies = test_data['Company'].unique()

# Calculate accuracy for each company
for company in companies:
    company_data = test_data[test_data['Company'] == company]
    accuracy = accuracy_score(company_data['target'], company_data['predictions'])
    results.append((company, accuracy))

# Convert the results to a DataFrame for better visualization
results_df = pd.DataFrame(results, columns=['Company', 'Accuracy'])

print(results_df)
```
